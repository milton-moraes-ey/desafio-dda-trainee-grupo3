```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='ticks', color_codes=True)
```

```python
#Carregamento de arquivo base com os dados dos Herois, seus Universos e Poderes

raw_data = pd.read_csv('HeroAnalysisData.csv')
raw_data.head()
```

```python
#Retirada da coluna com links de imagens

raw_data.drop(['SRC IMG'], axis='columns', inplace=True)
raw_data.head()
```

```python
#Retirada de espaços do começo das strings da coluna 'Universe'

raw_data['Universe'] = raw_data['Universe'].str.lstrip()
```

```python
#exportação dos dados sem espaço na coluna 'Universe' para uso no banco de dados:

raw_data.to_csv('UniverseSemEspaços.csv', index=False)
```

```python
#Transformação da coluna SuperPowers em um array

raw_data['SuperPowers_array'] = raw_data['SuperPowers'].str.split(',')
raw_data.head()
```

```python
#Transformação da coluna SuperPowers_array em Dummies

raw_data1 = (
    raw_data['SuperPowers_array'].explode().str.get_dummies().sum(level=0)
)

raw_data1 = raw_data.join(raw_data1)

raw_data1
```

```python
#Universos mais frequentes na base

raw_data1['Universe'].value_counts()
```

```python
#Transformação dos Dummies de poder

Powers = raw_data1.iloc[:,4:]
```

```python
#Soma de todos os Poderes

total_powers = Powers.sum(axis=None, skipna=True, level=None, numeric_only=None, min_count=0)
total_powers
```

```python
#Poderes mais frequentes na base

total_powers.sort_values(ascending = False).head()
```

```python
#Poderes menos frequentes na base

total_powers.sort_values(ascending = False).tail(38)
```

```python
#Universo isolado em um dataframe
Universo = raw_data1.iloc[:,1:2]
Universo
```

```python
#Concat de Universos e Poderes
Universo_Poder = pd.concat([Universo, Powers], axis=1)
Universo_Poder
```

```python
#Soma de todos os poderes por Universo
Universo_Poder.sum(axis=1, numeric_only=True)
```

```python
#Verificação da maior incidência de Poderes no Universo que mais aparece na base
filtro = (Universo_Poder['Universe'] == 'Prime Marvel Universe')
```

```python
MarvelPowers = Universo_Poder[filtro].sum(axis=None, skipna=True, level=None, numeric_only=None, min_count=0)
MarvelPowers
```
